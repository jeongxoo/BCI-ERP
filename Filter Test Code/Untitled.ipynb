```python
temp = pd.read_csv('0606_정수_rd/Rawdata10.txt',delimiter='\t',encoding='cp949')
e_fp1 = temp['EEG_Fp1']
```

```python
a = e_fp1[0:50]
```

```python
x = range(0,len(a))
Fs = 512
# lowcut = 0.5 #0.5
# highcut = 30 #30
# m_lpf = butter_bandpass_filter(e_fp1,lowcut, highcut, Fs, order=5)


# 그래프 그리는 부분
N = len(a)
k = np.arange(N)

T = N/Fs
freq = k / T
freq = freq[range(int(N/2))]

yfft = np.fft.fft(a)
yf = yfft / N
yf = yf[range(int(N/2))]

# plt.axhline(color='black')
plt.plot(x, a, 'b', label='origin')
plt.show()

plt.plot(freq,abs(yf),'gray')
plt.axvline(x = 30, color = 'red', alpha = 0.2)
plt.show()
```
